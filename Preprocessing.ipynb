```python
def punctuation_removal(sentences):
    processed_sentences = []
    for sent in sentences:
        processed = "".join([i for i in sent if i not in string.punctuation])
        processed = processed.strip()
        processed_sentences.append(processed)
    return processed_sentences
```

```python
train_sents = punctuation_removal(train_sents)
train_sents
```

```python
def name_removal(sentences):
    processed_sentences = []
    for sent in sentences:
        sent = sent.split(" ")
        for i in range(len(sent)):
            word = sent[i]
            if word.startswith("wzj"):
                sent[i] = "<name>"
        sent  = " ".join(sent)
        processed_sentences.append(sent)
    return processed_sentences
train_sents = name_removal(train_sents)
```

```python
def build_dataset(base_directory, save_path):
    sentences_file = base_directory + "/sents.txt"
    sentiments_file = base_directory + "/sentiments.txt"
    
    with open(sentences_file, "r", encoding = "utf8") as f1:
        sentences = [sentence.strip() for sentence in f1.readlines()]
    
    with open(sentiments_file, "r", encoding = "utf8") as f2:
        sentiments = [int(sentiment.strip()) for sentiment in f2.readlines()]
    
    sentences = punctuation_removal(sentences)
    sentences = word_segmentation(sentences)
    sentences = name_removal(sentences)
    
    df = pd.DataFrame()
    df["sentences"] = sentences
    df['sentiments'] = sentiments
    df.to_csv(save_path, index = False, sep = "|")
    return df
```

```python
train_dir = "./Data/train"
train_save_path = "./Data/processed/train.csv"

validation_dir = "./Data/dev"
validation_save_path = "./Data/processed/validation.csv"

test_dir = "./Data/dev"
test_save_path = "./Data/processed/test.csv"


train_df = build_dataset(train_dir, save_path  = train_save_path)
validation_df = build_dataset(validation_dir, save_path  = validation_save_path)
test_df = build_dataset(test_dir, save_path  = test_save_path)
```
